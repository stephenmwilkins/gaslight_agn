```python
import numpy as np
import matplotlib.pyplot as plt
from synthesizer.photoionisation import cloudy23
from astropy.io import ascii
from unyt import cm
```

```python
models = ['blr_default/1', 'blr_no-xrays/1', 'blr_50ev-cut/1']
models = ['blr_default/1', 'blr_no-Fe/1']
models = ['blr_default/1', 'blr_no-collisional-excitation/1']
```

## Ionisation structure

```python

zone = 'HII'

for model in models:
    data = ascii.read(f'{model}.ovr')
    depth = data['depth'] * cm
    plt.plot(depth.to('pc').value, data[zone], label=model)

plt.legend()
plt.xscale('log')
plt.ylim(-0.1, 1.2)
plt.xlim(1E-8, 1E-3)
plt.xlabel('depth/pc')
plt.show()
```
